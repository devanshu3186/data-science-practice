```python
import json
def load_data(filename):
    with open(filename, "r") as file:
        return json.load(file)
    
def people_you_may_know(user_id, data):
    #Finding every user friends
    user_friends={}
    for user in data['users']:
        user_friends[user['id']]=set(user['friends'])
    if user_id not in user_friends:
        return []
    #Finding users direct friend
    direct_friend=user_friends[user_id]
    suggestions={}
    for friend in direct_friend:
        for mutual in user_friends[friend]:
            if mutual!=user_id and mutual not in direct_friend:
                #count mutual firend
                suggestions[mutual]=suggestions.get(mutual,0)+1
    sorted_suggestion=sorted(suggestions.items(), key=lambda x: x[1], reverse=True)
    return [user_id for user_id, _ in sorted_suggestion]
    
    
    
data=load_data("massive_data.json")
user_id=10

recc=people_you_may_know(user_id, data)
print(recc)

    
```
