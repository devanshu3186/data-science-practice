Loads the json data:

```python
def load_data(filename):
    with open(filename, "r") as f:
        data = json.load(f)
    return data
```

```python
data = load_data("data.json")
```

Write a function to display user data and their connections:

```python
def display_user(data):
    print("User and thier connections\n")
    for user in data['users']:
          print(f"ID-{user['id']}: {user['name']} is friends with: {user['friends']} and liked pages are {user['liked_pages']}")
    print("\nPage Information\n")
    for page in data['pages']:
        print(f"{page['id']}: {page['name']} ")
display_user(data)
          
```
