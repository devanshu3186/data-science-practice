```python
import pandas as pd
```

```python
employees = pd.DataFrame({
    "EmpID": [1, 2, 3],
    "Name": ["Alice", "Bob", "Charlie"],
    "DeptID": [10, 20, 30]
})
employees
```

```python
departments = pd.DataFrame({
    "DeptID": [10, 20, 40],
    "DeptName": ["HR", "Engineering", "Marketing"]
})
departments
```

```python
pd.merge(employees, departments, on="DeptID") #Inner join
```

```python
pd.merge(employees, departments, on="DeptID", how="left") #Left join: Complete left table
```

```python
pd.merge(employees, departments, on="DeptID", how="right") #Right join: Complete right table
```

```python
pd.merge(employees, departments, on="DeptID", how="outer") #Outer join: Complete both tables
```

```python
df1 = pd.DataFrame({"Name": ["Alice", "Bob"]})
df2 = pd.DataFrame({"Name": ["Charlie", "David"]})
```

```python
pd.concat([df1, df2]) #concat vertically (Add Rows)
```

```python
pd.concat([df1,df2], axis=1) #Concat Horizontally (Add Columns)
```
