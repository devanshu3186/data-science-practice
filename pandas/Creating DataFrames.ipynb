# Creating DataFrame

```python
import pandas as pd
```

## From list of lists

```python
data=[["Harry", 89],["Carry", 79],["Rohan", 67]]
data
```

```python
df=pd.DataFrame(data,columns=["Name","Marks"])
df
```

## From Dictionary of Lists

```python
# Creating DataFrame Using 
data={"Name":["Atul","Bheem", "Kalia", "John"],
     "Marks":[24, 67, 54, 90]
     }
data
```

```python
df=pd.DataFrame(data)
df
```

## From NumPy Arrays

```python
# Creating DataFrame Using NumPy Arrays
import numpy as np
arr= np.array([[12,34,56],[57,47,98]])
arr
```

```python
df=pd.DataFrame(arr, columns=["a","b","c"])
df
```

## From Excel Files

```python
df=pd.read_excel("data1.xlsx")
df
```

# Creating DataFrame From CSV Files
