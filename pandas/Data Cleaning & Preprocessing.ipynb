```python
df2=df.dropna().copy()
```

```python
df2["Age"]= df2["Age"].astype(int)
df2
```

```python
df2.info()
```

```python
df2["Join Date"] = pd.to_datetime(df2["Join Date"], format="%d-%m-%Y")
```

```python
df2.info()
```

```python
df2["Age Group"]=df2["Age"].apply(lambda age: "Adult" if age>18 else "Minor")  # .apply() → Apply any function to rows or columns
```

```python
df2
```

```python
df3=df2.copy()
df3["Gender"] = df3["Gender"].str.strip()
gender_map = {"M": "Male", "F": "Female", "O": "Others"}
df3["Gender"] = df3["Gender"].map(gender_map)              # .map() → Element-wise mapping for Series
```

```python
df3
```

```python
df3["City"]=df3["City"].replace({"New York": "NYC", "Mumbai": "Bombay"})   # .replace() → Replace specific values
df3
```
