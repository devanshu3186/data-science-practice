```python
import pandas as pd

df=pd.read_csv("actorfilms.csv")
df
```

```python
df["Actor"] #Single column as Series
```

```python
df[["Actor","Film"]] #Multiple columns as DataFrame
```

```python
df.loc[0]  # First row (by label)
```

```python
df.iloc[0] # First row (by position)
```

```python
df.loc[0, "Actor"]        # Value at row 0, column 'Actor'
```

```python
df.iloc[0, 2]            # Value at row 0, column at index 2
```

```python
df.loc[35:40, ["Actor", "Film", "Rating"]] # Rows 35 to 40, selected columns
```

```python
df.iloc[35:40, [0,2,5]] # Rows and cols by index position
```

```python
df.at[0, "Actor"]       # Fast label-based access
```

```python
df.iat[0, 2]           # Fast position-based access
```

```python
df[(df["Rating"]>7.5) & (df["Year"]>=2019)]
```

```python
df.query("Year > 1995 and Rating == 8.5") #Query by default returns a copy
```

```python
df.query("Votes > 1000000 and Rating >= 8")
```

### loc, iloc and other filtering techniques(except query) always return a view and changes made to them may effect the original DataFrame.
### So, always use .copy() function if you want to further manipulate your filtered data without your original data being manipulated.
### EX: df2 = df[["Actor","Film"]].copy()
