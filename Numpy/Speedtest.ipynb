```python
#Python Zip Explained.
l1=[2,4,1]
l2=[3,2,6]
list(zip(l1,l2))
```

```python
#Using Python List.
import numpy as np
import time

size=1_000_000_00
l1=list(range(size))
l2=list(range(size))

start=time.time()
add=[x+y for x,y in zip(l1,l2)]
end=time.time()
print(add[0:10])
print("Python list addition time: ",end - start)
```

```python
#Using numpy Array.
import numpy as np
import time

size=1_000_000_00
l1=np.array(l1)
l2=np.array(l2)

start=time.time()
add=l1+l2
end=time.time()
print(add[0:10])
print("Numpy array addition time: ",end - start)
```

```python
import sys

list_data = list(range(1000))
numpy_data = np.array(list_data)
total = sys.getsizeof(list_data)

for item in list_data:
    total += sys.getsizeof(item)
print("Python list size:", total, "bytes")
print("NumPy array size:", numpy_data.nbytes, "bytes")
```

```python
# Python list (loop-based)
list1=[1,2,3,4]
start=time.time()
list_squares = [x ** 2 for x in list1]
end=time.time()
print("Time taken: ",end-start)
list_squares
```
