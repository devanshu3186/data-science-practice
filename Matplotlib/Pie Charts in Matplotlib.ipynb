```python
import matplotlib.pyplot as plt
plt.rcParams['figure.figsize'] = (6.4, 4.8)
plt.style.use("ggplot")
```

```python
labels = ["Sachin", "Sehwag", "Kohli", "Yuvraj"]
runs = [18000, 8000, 12000, 9500]
colours=["#FF0000","Green", "Blue", "yellow"]
 
plt.title("Career Runs of Indian Batsmen")
plt.pie(
    runs, 
    labels=labels, 
    colors=colours, 
    wedgeprops={'edgecolor': 'black', 'linewidth': 3, 'linestyle': '-', 'fill':True}
)
plt.tight_layout()
```

```python
labels = ["Sachin", "Sehwag", "Kohli", "Yuvraj"]
runs = [18000, 8000, 12000, 9500]
colors=["#FF0000","Green", "Blue", "yellow"]
explode=[0.1,0,0,0]
plt.pie(runs, labels=labels, colors=colors, explode=explode, shadow=True, startangle= -30, autopct='%1.1f%%')
plt.tight_layout()
```

```python
import matplotlib.pyplot as plt

languages = ['Python', 'Java', 'C++', 'JavaScript', 'C#', 'Ruby', 'Go', 'Rust', 'Swift', 'PHP']
usage = [30, 20, 10, 10, 7, 5, 4, 3, 2, 1]

# Keep the top 5 languages
top_languages = languages[:5]
top_usage = usage[:5]

# Combine the remaining into "Others"
others_usage = sum(usage[5:])

top_languages.append("Others")
top_usage.append(others_usage)

plt.pie(top_usage,
        labels=top_languages,
        autopct='%1.1f%%',
        startangle=90)

plt.title("Programming Language Usage")
plt.show()
```
