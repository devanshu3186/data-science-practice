```python
import matplotlib.pyplot as plt

years = [1990, 1992, 1994, 1996, 1998, 2000, 2003, 2005, 2007, 2010]
runs =  [500, 700, 1100, 1500, 1800, 1200, 1700, 1300, 900, 1500]

plt.bar(years, runs)
plt.xlabel("Year")
plt.ylabel("Runs Scored")
plt.title("Sachin Tendulkar's Yearly Runs")
```

```python
import numpy as np

sachin = [500, 700, 1100, 1500, 1800, 1200, 1700, 1300, 900, 1500]
sehwag = [0, 200, 900, 1400, 1600, 1800, 1500, 1100, 800, 0]
kohli  = [0, 0, 500, 800, 1100, 1300, 1500, 1800, 1900, 2100]
years = [1990, 1992, 1994, 1996, 1998, 2000, 2003, 2005, 2007, 2010]

x = np.arange(len(years))  # index positions
width = 0.25

plt.bar(x - width, sachin, width=width, label="Sachin")
plt.bar(x, sehwag, width=width, label="Sehwag")
plt.bar(x + width , kohli, width=width, label="Kohli")

plt.xlabel("Year")
plt.ylabel("Runs")
plt.title("Run Comparison")
plt.xticks(x, years)  # Show actual year instead of 0,1,2,...
plt.legend()
plt.tight_layout()
```

```python
players = ["Sachin", "Sehwag", "Kohli", "Yuvraj"]
runs_5yrs = [500+700+1100+1500+1800, 0+200+900+1400+1600, 0+0+500+800+1100, 300+600+800+1100+900]

plt.barh(players, runs_5yrs, color="skyblue")
plt.xlabel("Total Runs in First 5 Years")
plt.title("First 5-Year Performance of Indian Batsmen")
plt.tight_layout()
```

```python
players = ["Sachin", "Sehwag", "Kohli"]
runs = [1500, 1200, 1800]
 
plt.bar(players, runs, color="skyblue")
 
# Add labels on top of bars
for i in range(len(players)):
    plt.text(i, runs[i] + 30, str(runs[i]), ha='center')
 
plt.ylabel("Runs")
plt.title("Runs Scored by Players")
plt.tight_layout()
plt.show()
```

```python
import pandas as pd

df = pd.DataFrame({
    'Store A': [25, 40, 30, 55],
    'Store B': [30, 35, 45, 50]
}, index=['Apples', 'Bananas', 'Oranges', 'Grapes'])

# Pandas automatically clusters columns side-by-side
df.plot(kind='barh', color=['skyblue', 'orange'])
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Sample data
categories = ['Apples', 'Bananas', 'Oranges', 'Grapes']
store_a = [25, 40, 30, 55]
store_b = [30, 35, 45, 50]

# Generate index array for Y positions
y = np.arange(len(categories))
height = 0.35  # Thickness of each bar

fig, ax = plt.subplots()

# Plot first dataset shifted up, second dataset shifted down
rects1 = ax.barh(y - height/2, store_a, height=height, label='Store A', color='skyblue')
rects2 = ax.barh(y + height/2 , store_b, height=height, label='Store B', color='orange')

# Add labels, titles, and formatting
ax.set_ylabel('Fruit Types')
ax.set_xlabel('Quantity Sold')
ax.set_title('Sales Comparison by Store')
ax.set_yticks(y)
ax.set_yticklabels(categories)
ax.legend()

plt.tight_layout()
plt.show()
```
