```python
with open("initialdata.txt","r")as f:
    data=f.read()
    
chunks=data.split("\n\n")
chunks = [c for c in chunks if len(c)>3]
def parse_chunk(chunk):
    chunk=chunk.strip()
    sep_chunk=chunk.split("\n")
    username=sep_chunk[0]
    no_of_posts=sep_chunk[1]
    no_of_followers=sep_chunk[2]
    no_of_following=sep_chunk[3]
    name=sep_chunk[4]
    if (len(sep_chunk) > 5):
        type_of_page=sep_chunk[5]
        bio="\n".join(sep_chunk[6:])
    else:
        type_of_page = "Unknown"
        bio = ""
    # print(sep_chunk)
    return {"username":username, "no_of_post": no_of_posts, "no_of_followers":no_of_followers, "no_of_following":no_of_following, "name":name, "type_of_page":type_of_page, "bio":bio}
    
# parse_chunk(chunks[0])
```

```python
all_chunks=[]
for chunk in chunks:
    parsed_chunks=parse_chunk(chunk)
    all_chunks.append(parsed_chunks)
print(all_chunks)
    
```
